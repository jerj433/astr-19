# Session 6: sin(x) and cos(x) functions
This notebook defines functions for sin(x) and cos(x), then tabulates them from x=0 to x=2 using 1000 points.

```python
import numpy as np
```

```python
def my_sin(x):
    """Return sin(x)."""
    return np.sin(x)
```

## cos(x)
This function returns the cosine of x.

```python
def my_cos(x):
    """Return cos(x)."""
    return np.cos(x)
```

```python
x_vals = np.linspace(0.0, 2.0, 1000)
sin_vals = my_sin(x_vals)
cos_vals = my_cos(x_vals)
```

```python
print("x\t\tsin(x)\t\tcos(x)")
for i in range(10):
    x = x_vals[i]
    print(f"{x:.6f}\t{sin_vals[i]:.6f}\t{cos_vals[i]:.6f}")
```
